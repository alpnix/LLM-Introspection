## Experiment 1

```python
from ollama import embeddings
import numpy as np

all_caps = "HI, HOW ARE YOU"
no_caps = "hi, how are you"

caps_vector = embeddings(model="llama2", prompt=all_caps)
no_caps_vector = embeddings(model="llama2", prompt=no_caps)
```

```python
def cosine_similarity(v1, v2): 
    v1 = np.array(v1)
    v2 = np.array(v2)
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def vector_difference(v1, v2): 
    v1 = np.array(v1)
    v2 = np.array(v2)
    return v1 - v2
```

```python
caps_similarity = cosine_similarity(caps_vector.embedding, no_caps_vector.embedding)
caps_difference = vector_difference(caps_vector.embedding, no_caps_vector.embedding)
print(f"Cosine similarity: {caps_similarity}")
print(f"Vector difference: {caps_difference}")
```
